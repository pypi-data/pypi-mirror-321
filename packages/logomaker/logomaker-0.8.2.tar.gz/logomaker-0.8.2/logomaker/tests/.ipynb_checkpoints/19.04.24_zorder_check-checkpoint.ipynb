```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import sys
sys.path.append('../../')
import logomaker
```

```python
df = pd.read_csv('../figure/matrices/crp_negenergy_matrix.txt', 
                 delim_whitespace=True, 
                 index_col=0)
```

```python
fig, ax = plt.subplots(1,1,figsize=[10,2])

xs = np.linspace(0,25,1000)
ys = np.sin(xs)

ax.plot(xs,ys, zorder=1, linewidth=3)
logo = logomaker.Logo(df, ax=ax, zorder=0, 
                      draw_now=False)
logo.style_glyphs_in_sequence('.....TGTGA......TCACA.....', color='cyan', 
                              draw_now=False, zorder=2)
logo.draw()
```
