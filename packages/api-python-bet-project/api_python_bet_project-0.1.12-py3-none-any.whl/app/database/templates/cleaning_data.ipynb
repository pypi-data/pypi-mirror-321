# Cleaning data

#### Reading the csv

```python
# Define the filename for saving match data
filename = f'matches_{season}.csv'

# Reading the csv
df = pd.read_csv(filename)
```

#### Making day, time, day_year of the match 

```python
# Convert the 'date_of_match' column to datetime using the correct format
df['date_of_match'] = pd.to_datetime(df['date_of_match'], format="%Y-%m-%d")

# Attempt conversion with coercion for invalid formats
df['hour_of_the_match'] = pd.to_datetime(df['hour_of_the_match'], format="%H:%M", errors='coerce')
```

```python
# Extract day of the week from the date
df['day_of_the_week'] = df['date_of_match'].dt.dayofweek

# Convert the time to datetime and calculate the time in hours as a decimal
df['time_of_the_match'] = pd.to_datetime(df['hour_of_the_match'], format="%H:%M").apply(
    lambda x: x.hour + x.minute / 60
)

# Extract day of the year from the date
df['day_of_year'] = df['date_of_match'].dt.dayofyear
```
